# First Part: General Sales Analysis

• What is the distribution of sales by product category?
• What are the top-selling products?
• What is the sales trend over time? (by hour)

```python
sales_by_category = coffee.groupby("product_category")["transaction_qty"].sum().sort_values(ascending=False)
# Group sales by product category and sum the quantity sold

# Displaying the graph
plt.figure(figsize=(10, 3))  # Set the size of the chart
sales_by_category.plot(kind="bar", color='saddlebrown')  # Plot a bar chart, add color
# Creating the chart with thousands separator on the Y axis
plt.title("Sales distribution by product category", fontsize=15, fontweight="semibold", color="#AD7B51") # Add a title
plt.xlabel("Product Category") # Add a label for the X axis
plt.ylabel("Quantity sold") # Add a label for the Y axis
y_ticks = [0, 20000, 40000, 60000, 80000]  
plt.yticks(y_ticks, [f"{y:,}" for y in y_ticks], fontsize=10)  
# Keep current scale and add thousands separator
plt.xticks(rotation=65, fontsize=10)  # Tilt labels for better readability
# Rotate labels for better readability
plt.show() # Show graph
```

```python
top_products = coffee.groupby("product_detail")["transaction_qty"].sum().sort_values(ascending=False).head(10)
# Group sales by product and sum the quantity sold

colors = ["#D4A373" if x < max(top_products) else "#8B4513" for x in top_products]
# Define a color palette (gradient highlighting the top 1 product)

min_x = min(top_products) * 0.9  # We start a little before the smallest value
max_x = max(top_products) * 1.05  # We add a margin of 5% at most to avoid crushing
# Set dynamic limits to shorten the X axis

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.barh(top_products.index, top_products, color=colors, edgecolor="black", alpha=0.9)
for bar in bars:
    ax.text(bar.get_width() + (max_x * 0.01), bar.get_y() + bar.get_height()/2, 
            str(int(bar.get_width())), ha='left', va='center', 
            fontsize=12, fontweight='semibold', color='#D2B48C')
ax.set_xlim(min_x, max_x)
# Set X limits to not display the whole scale unnecessarily
ax.set_title("Top 10 Best Selling Products", fontsize=15, fontweight="semibold", color="#AD7B51")
ax.set_xlabel("Quantity sold", fontsize=12, fontweight="medium", color="black")
ax.set_ylabel("Product", fontsize=12, fontweight="medium", color="black")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_linewidth(1.2)
ax.spines["left"].set_linewidth(1.2)
# Remove unnecessary borders for a smoother design
ax.invert_yaxis()
# Invert the Y axis to put the best product on top
ax.grid(False)
# Remove the grid for a cleaner effect
plt.show()
```

# Second Part: Store Analysis

• Which stores have the most sales?
• Are there differences in sales depending on the location of the stores?

```python
sales_by_location = coffee.groupby("store_location")["transaction_qty"].sum().sort_values(ascending=False)
# Group sales by store location

colors = ["#8B4513", "#A0522D", "#D2691E", "#CD853F", "#DEB887", "#F4A460", "#D2B48C"]
# Define a brown color palette, a harmonious design with colors adapted to the world of coffee

explode = [0.1 if i == sales_by_location.idxmax() else 0 for i in sales_by_location.index]
# Set explode to detach the store that makes the most sales

plt.figure(figsize=(10, 6))
plt.pie(sales_by_location, labels=sales_by_location.index, autopct='%1.1f%%', 
        colors=colors, startangle=140, explode=explode, wedgeprops={"edgecolor": "black"})
plt.title("Distribution of sales by store location", fontsize=15, fontweight="semibold", color="#AD7B51", pad=30)
plt.axis("equal")
# Ensure the pie chart remains circular
plt.show()
```

Conclusion of the store analysis:

We found that some stores dominate sales, while others record lower volumes. This difference can be influenced by several factors, including the location of the store, the number of customers, the popularity of the products offered or the quality of the service.

• The use of a pie chart with an "explode" effect made it possible to highlight the store generating the most sales, thus presenting its key role in the network ('Lower Manhattan'). These results reveal several avenues for improvement. It would be relevant to analyze the reasons for the success of the best stores in order to adopt these good practices in the other stores, to adjust stocks according to the performance of the different locations and to implement targeted marketing strategies according to the specificities of each store. # Third Part: Revenue Analysis

• What is the total revenue generated by product category?
• Which products are generating the most revenue?
• what is the average turnover per day of the week?
• Is there seasonality in coffee shop revenue?

# The “K” is actually an abbreviation for “kilo,” which means 1,000. 
So when you see “10K” on a data dashboard, it means “10,000.” Similarly, “100K” means “100,000.”

```python
coffee["revenue"] = coffee["transaction_qty"] * coffee["unit_price"]
revenue_by_category = coffee.groupby("product_category")["revenue"].sum().sort_values(ascending=False)
# Calculation of turnover by product category

plt.figure(figsize=(7, 6))
revenue_by_category.plot(kind="bar", color='saddlebrown')
# Create the graph with thousands separator and add the "k" on the Y axis
plt.title("Turnover by product category", fontsize=15, fontweight="semibold", color="#AD7B51")
plt.xlabel("Product Category")
plt.ylabel("Total income (k$)")  # Added "k$" in the label
y_ticks = [50000, 100000, 150000, 200000, 250000]  
plt.yticks(y_ticks, [f"{y//1000}k" for y in y_ticks], fontsize=10)  
plt.xticks(rotation=62, fontsize=10)
plt.figure(figsize=(7, 6))
plt.show()
```

```python
revenue_by_product = coffee.groupby("product_detail")["revenue"].sum().sort_values(ascending=False).head(10)
# Calculation of turnover by product (top 10)

max_revenue_product = revenue_by_product.idxmax()
# Determine the product that generates the most turnover

colors = ["saddlebrown" if product == max_revenue_product else "#D4A373" for product in revenue_by_product.index]
plt.figure(figsize=(12, 6))
revenue_by_product.plot(kind="barh", color=colors, edgecolor="black")
# Create the graph with an X-axis scale adjusted from 10,000 and display in "k"
plt.title("Top 10 Highest Revenue Generating Products", fontsize=15, fontweight="semibold", color="#AD7B51")
plt.xlabel("Total income (k$)", fontsize=12)
plt.ylabel("Product", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlim(10000, revenue_by_product.max() + 2000)
x_ticks = range(10000, int(revenue_by_product.max()) + 2000, 10000)
plt.xticks(x_ticks, [f"{x//1000}k" for x in x_ticks], fontsize=10)
# Adjust the X axis to start at 10,000 and format the values as "k"
for i, value in enumerate(revenue_by_product):
    plt.text(value + 100, i, f"{value/1000:.1f}k", va="center", fontsize=12, fontweight="semibold", color="#D2B48C")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

```python
coffee["transaction_date"] = pd.to_datetime(coffee["transaction_date"])
# Convert transaction_date column to datetime format

coffee["day_of_week"] = coffee["transaction_date"].dt.day_name()
# Extract the day of the week

coffee["revenue"] = coffee["transaction_qty"] * coffee["unit_price"]
# Calculate the turnover for each transaction

coffee_daily = coffee.groupby("transaction_date")["revenue"].sum().reset_index()
coffee_daily["day_of_week"] = coffee_daily["transaction_date"].dt.day_name()
# Calculate the total turnover per day

corrected_avg_revenue_per_day = coffee_daily.groupby("day_of_week")["revenue"].mean().reset_index()
# Calculate the average turnover per day of the week

order_days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
corrected_avg_revenue_per_day["day_of_week"] = pd.Categorical(
    corrected_avg_revenue_per_day["day_of_week"], categories=order_days, ordered=True
)
corrected_avg_revenue_per_day = corrected_avg_revenue_per_day.sort_values("day_of_week")
# Sort the days of the week in chronological order

max_revenue_day = corrected_avg_revenue_per_day["revenue"].idxmax()
max_day_name = corrected_avg_revenue_per_day.loc[max_revenue_day, "day_of_week"]
# Determine the day with the highest average turnover

colors = ["saddlebrown" if day == max_day_name else "#D4A373" for day in corrected_avg_revenue_per_day["day_of_week"]]
plt.figure(figsize=(10, 6))
plt.bar(corrected_avg_revenue_per_day["day_of_week"], corrected_avg_revenue_per_day["revenue"], edgecolor="saddlebrown", color=colors)
plt.title("Average turnover by day of the week", fontsize=15, fontweight="semibold", color="#AD7B51")
plt.xlabel("Day of the week", fontsize=12)
plt.ylabel("Average turnover ($)", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(range(3200, int(corrected_avg_revenue_per_day["revenue"].max()) + 200, 200), fontsize=10)
plt.ylim(3200, corrected_avg_revenue_per_day["revenue"].max() + 200)
# Create the chart with intervals of 200 on the Y axis starting from 3200
for i, value in enumerate(corrected_avg_revenue_per_day["revenue"]):
    plt.text(i, value + 25, f"{value:.0f}", ha="center", fontsize=12, fontweight="semibold", color="#D2B48C")
plt.grid(axis="y", linestyle=":", alpha=0.5)
plt.show()
```

```python
coffee["transaction_date"] = pd.to_datetime(coffee["transaction_date"])
# Convert transaction_date column to datetime

coffee["revenue"] = coffee["transaction_qty"] * coffee["unit_price"]
# Add income column

revenue_by_category = coffee.groupby("product_category")["revenue"].sum().sort_values(ascending=False)
# Total revenue generated by product category

top_revenue_products = coffee.groupby("product_detail")["revenue"].sum().sort_values(ascending=False).head(10)
# Products generating the most revenue

coffee["month"] = coffee["transaction_date"].dt.month  # Extract the month number
coffee["month_name"] = coffee["transaction_date"].dt.strftime("%B") # Extract the name of the month
revenue_by_month = coffee.groupby("month_name")["revenue"].sum()
# Extract month and year then group by month

months_order = [
    "January", "February", "March", "April", "May", "June", "July",
    "August", "September", "October", "November", "December"
]
revenue_by_month = revenue_by_month.reindex(months_order)
# Sort months in chronological order
plt.figure(figsize=(12, 6))
plt.plot(revenue_by_month.index, revenue_by_month.values, marker='*', linestyle='-', color="saddlebrown")
plt.title("Monthly evolution of turnover (Seasonality)", fontsize=15, fontweight="semibold", color="#AD7B51")
plt.xlabel("Month", fontsize=12)
plt.ylabel("Turnover (K$)", fontsize=12)
plt.xticks(rotation=45)
# Format the Y axis with "k"
y_max = int(revenue_by_month.max())
y_ticks = range(0, y_max + 50000, 50000)  # Intervals of 50,000
plt.yticks(y_ticks, [f"{y//1000}k" for y in y_ticks], fontsize=10)
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

Conclusion of the revenue analysis:

• Revenue analysis allowed us to better understand which product categories and items generate the most revenue, as well as identify the existence of seasonal trends influencing coffee shop sales and profitability.
We found that some product categories largely dominate revenue, while others have a more limited impact. The most praiseworthy products are not necessarily the best-selling ones.
Analysis of the seasonality of revenue reveals significant fluctuations over the months, with increased consumption of hot drinks in winter, the influence of holiday periods, and the impact of promotions and special events.

• These observations highlight several optimization levers:
✔️ Maximize sales of the most profitable products by highlighting those that generate the most revenue, through special offers or loyalty campaigns.
✔️ Balance offers and adjust the pricing strategy by taking into account the most requested products and optimizing profit margins.
✔️ Adapt inventory and marketing campaigns to seasonal variations, in order to maximize sales during periods of high demand and avoid losses during periods of low demand.

• In conclusion, this analysis highlights the importance of tracking revenue not only based on sales volume, but also by taking into account prices and seasonal variations. A strategic approach based on these trends allows you to optimize the overall profitability of coffee shops, anticipate market fluctuations and adjust business operations for sustainable growth.

```python
coffee["total_price"] = coffee["transaction_qty"] * coffee["unit_price"]
transaction_totals = coffee.groupby("transaction_id")["total_price"].sum()
# calculate total spent per transaction

bins = [0, 2, 4, 6, 8, 10, 15, 20, 30, 50] # Price ranges
labels = ["0-2($)", "2-4($)", "4-6($)", "6-8($)", "8-10($)", "10-15($)", "15-20($)", "20-30($)", "30-50($)"]
# Define price ranges to better visualize the distribution of expenses

transaction_totals_binned = pd.cut(transaction_totals, bins=bins, labels=labels, right=False)
# Categorize each transaction into a range

price_distribution = transaction_totals_binned.value_counts().sort_index()
# Count the number of transactions per price interval

fig, ax1 = plt.subplots(figsize=(10, 4))
price_distribution.plot(kind="bar", color="#D4A373", edgecolor="saddlebrown", alpha=0.8, ax=ax1)

# Add a more stylized curve on the same graph
ax2 = ax1.twinx()  # Create a second Y axis for the curve
ax2.plot(price_distribution.index, price_distribution.values, marker=".", linestyle="-", 
         color="#6F4E37", linewidth=3, markersize=8, markerfacecolor="white", markeredgewidth=2, label="Transaction curve")
ax1.set_title("Distribution of transactions by price range ($)", fontsize=15, fontweight="semibold", color="#AD7B51")
ax1.set_xlabel("Mid-price range ($)", fontsize=12)
ax1.set_ylabel("Number of transactions (bar)", fontsize=12, color="black")
ax2.set_ylabel("Number of transactions (curve)", fontsize=12, color="saddlebrown")
y_ticks = [0, 10000, 30000, 50000, 70000]
ax1.set_yticks(y_ticks)
ax1.set_yticklabels([f"{y:,}" for y in y_ticks], fontsize=10)  # Format with thousands separator
# Apply the same scale to the curve for perfect alignment
ax2.set_yticks(y_ticks)
ax2.set_yticklabels([f"{y:,}" for y in y_ticks], fontsize=10, color="saddlebrown")
ax1.set_xticklabels(price_distribution.index, rotation=45, fontsize=10)
ax1.grid(axis="y", linestyle=":", alpha=0.7)
# Add a grid to improve readability
ax2.fill_between(price_distribution.index, price_distribution.values, color="saddlebrown", alpha=0.2)
# Add a shadow effect under the curve for a better visual effect
ax2.legend(loc="upper right", fontsize=11, frameon=False)
# Show legend with more elegant style
plt.show()
```

Over a year (240 working days):
4.69 × 240 = 1,125.60$ per year
A regular customer can spend more than 1,100$ per year on coffee alone, which is a significant budget.

In Tel Aviv, it is common for workers to buy several coffees per day, especially in the morning and after lunch.
